```python
from geopy.geocoders import Nominatim
from google.colab import files
import folium
from folium.plugins import MiniMap
from branca.element import Figure

campus_list = ["Campus Omar Dengo",
              "Campus Benjamín Núñez",
              "Sede Interuniversitaria",
              "Campus Sarapiquí",
              "Campus Coto",
              "Campus Pérez Zeledón",
              "Campus Liberia",
              "Campus Nicoya"]

geolocator=Nominatim(user_agent="steamcoder")
my_map = folium.Map(location=[9.97389895,-84.1280985519902])


# Define a dictionary to map campus names to colors
campus_colors = {
    "Campus Omar Dengo": "red",
    "Campus Benjamín Núñez": "blue",
    "Sede Interuniversitaria": "green",
    "Campus Sarapiquí": "purple",
    "Campus Coto": "orange",
    "Campus Pérez Zeledón": "darkred",
    "Campus Liberia": "lightred",
    "Campus Nicoya": "beige"
}

# Define a dictionary to map campus names to icons
campus_icons = {
    "Campus Omar Dengo": "info-sign",
    "Campus Benjamín Núñez": "book",
    "Sede Interuniversitaria": "globe",
    "Campus Sarapiquí": "tree",
    "Campus Coto": "home",
    "Campus Pérez Zeledón": "star",
    "Campus Liberia": "plane",
    "Campus Nicoya": "anchor"
}

# Iterate through the campus list and add markers with custom colors and icons
for campus in campus_list:
    location = geolocator.geocode(campus)
    if location:
        color = campus_colors.get(campus, "blue")  # Default to blue if color not found
        icon = campus_icons.get(campus, "info-sign")  # Default to info-sign if icon not found
        folium.Marker(
            [location.latitude, location.longitude],
            popup=campus,
            icon=folium.Icon(color=color, icon=icon),
        ).add_to(my_map)

my_map
```
