```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame

# Function to check if a genre string contains "Adventure"
def contains_adventure(genre_string):
    if pd.isnull(genre_string):  # Handle missing values
        return False
    genres = str(genre_string).split(',')  # Split genres if multiple
    return any("Adventure" in genre for genre in genres)

# Apply the function to create a new column indicating if a movie is an "Adventure" movie
data['is_adventure'] = data['genres'].apply(contains_adventure)

# Filter for movies classified as "Adventure"
adventure_movies = data[data['is_adventure']]

# Get top 5 production companies for "Adventure" based on the count of released movies
top_5_companies = adventure_movies['production_companies'].value_counts().head(5).index

# Filter DataFrame for top 5 companies and "Adventure" movies
filtered_df = adventure_movies[adventure_movies['production_companies'].isin(top_5_companies)]

# Create a contingency table for "Adventure" movies only
contingency_table = pd.crosstab(filtered_df['production_companies'], filtered_df['is_adventure'])

# Rename the column to "Adventure"
contingency_table = contingency_table.rename(columns={True: 'Adventure'})

# Create heatmap with custom colormap
plt.figure(figsize=(10, 10))
sns.heatmap(contingency_table, annot=True, cmap='RdBu_r', fmt='d', cbar=True)
plt.title('Top 5 Production Companies for Adventure Movies (Including Combinations)')
plt.xlabel('Genre')
plt.ylabel('Production Company')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

contingency_table = contingency_table.sort_values(by=['Adventure'], ascending=False)

# Create heatmap with custom colormap
plt.figure(figsize=(10, 10))
sns.heatmap(contingency_table, annot=True, cmap='RdBu_r', fmt='d', cbar=True)
plt.title('Top 5 Production Companies for Adventure Movies (Ascending Order)')
plt.xlabel('Genre')
plt.ylabel('Production Company')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Cargar el dataset
file_path = '/content/drive/MyDrive/TMDB_movie_dataset_v11.csv'
data = pd.read_csv(file_path)

# Convertir la columna 'release_date' a formato de fecha
data['release_date'] = pd.to_datetime(data['release_date'], errors='coerce')

# Filtrar películas recientes (últimos 10 años)
recent_data = data[data['release_date'].dt.year >= 2010]

# Filtrar películas con datos válidos de presupuesto y revenue
data_filtered = recent_data[(recent_data['budget'] > 0) & (recent_data['revenue'] > 0)]

# Seleccionar los 15 géneros más populares
top_15_genres = data_filtered['genres'].value_counts().head(15).index
top_genre_data = data_filtered[data_filtered['genres'].isin(top_15_genres)]

# Crear un dataframe que agrupe por género y calcule el promedio de presupuesto, revenue y popularidad
grouped_data = top_genre_data.groupby('genres').agg({'budget': 'mean', 'revenue': 'mean', 'popularity': 'mean'}).reset_index()


# Preparar los datos para el heatmap
heatmap_data = grouped_data.set_index('genres')[['budget', 'revenue', 'popularity']]

# Normalizar los datos
scaler = MinMaxScaler()
heatmap_data_scaled = scaler.fit_transform(heatmap_data)
heatmap_data_scaled = pd.DataFrame(heatmap_data_scaled, columns=heatmap_data.columns, index=heatmap_data.index)

# nombre de columnas personalizado
heatmap_data_scaled.columns = ['Presupuesto', 'Ganacias', 'Popularidad']

# Crear el heatmap con más personalización
plt.figure(figsize=(15, 8))
sns.heatmap(heatmap_data_scaled, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Mapa de Calor: Relación entre Budget, Revenue y popularidad por Género (2010-2023)', fontsize=16)
plt.xlabel('Métricas', fontsize=12)
plt.ylabel('Géneros', fontsize=12)

plt.show()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
